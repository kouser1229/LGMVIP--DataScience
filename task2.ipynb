```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
```

# Loading the image

```python
img=cv2.imread('bg2.png', cv2.IMREAD_COLOR)
```

```python
img
```

# Displaying image

```python
display(Image.fromarray(img))
```

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.savefig('temp.png')
plt.show()
```

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
display(Image.fromarray(img_gray))
```

```python
img_invert=cv2.bitwise_not(img_gray)
display(Image.fromarray(img_invert))
```

```python
img_smoothing=cv2.GaussianBlur(img_invert,(21,21),sigmaX=0,sigmaY=0)
display(Image.fromarray(img_smoothing))
```

```python
def vce(x,y):
    return cv2.divide(x,255-y,scale=256)
final_img=vce(img_gray,img_smoothing)
display(Image.fromarray(final_img))
```

# Now we got the pencil sketch of the image
